Dictionaries are unordered collections of items. They store data in key-value pairs. Keys must be unique and
immutable (e.g., strings, numbers, or tuples), while values can be of any type.

```python
#  creating dectionaries

dec = {1 : "mayur",}
print(type(dec))
print(dec[1])
```

```python
std = {"name " : "mayur" , "roll" : '1', "div" : "A"} # one key at one time 
print(std)
```

```python
# accesing elements 
std = {"name" : "mayur" , "roll" : '1', "div" : "A"} 
print(std["name"])
print("the roll no. is ", std["roll"])

# accesing using get() method 
print(std.get("name"))
print(std.get('last_name')) # if element is not present it give none '
print(std.get('last_name',"not Available")) # getting value which we provided '


```

```python
# modifying dectinaries elements 
# dect is mutable , so we can add or update or delete elements 
print(std)
std["name"] = "ram krishna" #changing elements 
print(std)
std["city"] = "nandurbar" # adding new element 
print(std)

del std["city"] # deleting elements 
print(std)
```

```python
# dictionaries methods 
keys = std.keys() # printing all keys
print(keys)
values = std.values() #printing all values
print(values)
items = std.items() #printing all items 
print(items)
```

```python
# shadow copy 

# methond 1  (it will take difffrent memory )
std_copy = std.copy()
print(std_copy)

# method 2  (it will take same memory it we change value in main dict the cangest will auto updated in copy dict)
std_copy2 = std 
print(std_copy2)
```

```python
# iterating over keys
for keys in std.keys() :
    print(keys)
```

```python
# iterating over values 
for values in std.values():
    print(values)
```

```python
# iterating over items 
for key,values in std.items():
    print(f"{keys} : {values}")
```

```python
# nested dect 
student= {
    "student1" : {"name" : "mayur","age" : 20},
    "student2" : {"name" : "om","age" : 18}
}
print(student)
```

```python
# accesing nested dect elements 
print(student['student1']['name'])
print(student['student2']['name'])
```

```python
# iterating over elements in nested dict 
for student_id,student_info in student.items():
    print(f"{student_id} : {student_info}")
    for keys ,values in student_info.items():
        print(f"{keys}: {values}")
```

```python
# dict comprehansion 
dect_new = {x : x**2 for x in range(5)}
print(dect_new)
```

```python
# conditional dict comprehansion
even = {x : x**2 for x in range (10) if x%2==0 }
print(even)
```

```python
# merge 2 dict 
dict1 = {"name" :"mayur" , "roll" :"1"}
dict2 = {"name" :"om" , "roll" :"2","age":15}
dict3 = {**dict1,**dict2}
print(dict3)
```
