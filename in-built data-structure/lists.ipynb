```python
# slicing
list_2 = [1,2,3,4,5,6,7,8,9,]
print(list_2)
print(list_2[1:])
print(list_2[:5])
print(list_2[2:6])
print(list_2[::2])
print(list_2[::-2])
print(list_2[::-1])
```

```python
# list in for loop 
for index,i in enumerate(list_2) :
    print(i,index)
```

```python
# list comprenhesion 
lst = []
for i in range(10) :
    lst.append(i**2)
print(lst)
```

```python
# list comprehension in one line 
# type 1 [expression for item in iteration]
lst1 = [i**2 for i in range(10)]
print(lst1)
```

```python
# list comprehension in one line 
# type 2 [expression for item in iteration if condition]
lst3 = [num for num in range(10) if num%2==0]
print(lst3)
#for i in range(10):
#   if i%2==0:
#       lst.append(i) 
# 

```

```python
# list comprehension in one line 
# type 3 [expression for 1 item in 1 iteration for 2 item in 2 iteration]
l = [1,2,3,4]
l1 = ['a','b','c','d']
lst_4 = [[i,j] for i in l for j in l1 ]
print(lst_4)

```

```python
l2 = [1,2,3,4]
l3 = ['a','b','c','d']
l4=[]
for i in  l2:
    for j in l3:
        l4.append([i,j])
print(l4)
```
