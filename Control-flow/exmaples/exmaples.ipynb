```python
amount = int(input("enter the amount "))
if amount <= 5000 :
    print("there is your final amount",amount)
elif amount >5000 and amount <=  7000:
    discount_amount = amount*5/100
    new_amount = amount-discount_amount 
    print("this is your discounted amount",new_amount)
elif amount >7000 and amount <=9000:
    discount_amount = amount*10/100
    new_amount = amount-discount_amount 
    print("this is your discounted amount",new_amount)
elif amount > 9000 :
    discount_amount = amount*20/100
    new_amount = amount-discount_amount 
    print("this is your discounted amount",new_amount)
else :
    print("amount you have enter is invalid")
    

```
