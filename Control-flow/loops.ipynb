```python
# for loop 
for i in range(1,10):
    print(i)
```

```python
# example 2 
for i in range(1,10,2) :
    print(i)
```

```python
# for-loop in string 
str = "mayuyur"
for i in str :
    print(i)
```

```python
# while loop 
i = 0 
while i<5 :
    print(i)
    i = i+1 
```

```python
# break
# break statement break the loop (come out side )
for i in range(1,10):
    if i ==5 :
        break 
    print (i)
```

```python
# continue 
## The continue statement skips the current iteration and continues with the next.
for i in range(1,10):
    if i%2==0:
        continue
    print(i)
```

```python
# pass 
# it do nothing (timepass)
for i in range (5):
    if(i==3):
        pass #no change will happen (when we do not want to write anything so we write pass statement)
    print(i)
```

```python
# nested loop 
# loop inside loop 
for i in range(3):
    for j in range(2):
        print(i,j)
```
