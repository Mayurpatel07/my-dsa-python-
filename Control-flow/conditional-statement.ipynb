```python
##if elif statement 

##Calculater exmaple

num1= int(input("enter num1"))
num2= int(input("enter num2"))

operator = str(input("select which type of operation you want to perform (Add,sub,mult,divide"))

if(operator == "+" ):
    print(num1 + num2)
elif(operator == "-"):
    print(num1 - num2)
elif(operator == "*"):
    print(num1 * num2)
elif(operator == "/"):
    print(num1 / num2)
else:
    print("wrong operation choosed retry again !! ")
```
