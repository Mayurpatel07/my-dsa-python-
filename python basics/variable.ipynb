```python
##veriable types 
age = 25               # integer
height = 5.9          # float
name = "John Doe"     # string 

print(type(age))
print(type(height))
print(type(name))
```

```python
##exmaple of type casting in dynamic way (python support dynamic)
age = 11.11
print(type(age))
print("value of age is",age)

age = int(age)
print(type(age))
print("value of new age is",age)
```
